Generating data description and finding interquartile range for all features

```python
import pandas as pd

data=pd.read_csv("/content/A1.csv.csv")
print("Data Description")
numeric_data=data.select_dtypes(include=['number'])
print(numeric_data.describe())
print("/n Interquartile range for each feature")
Q1=numeric_data.quantile(0.25)
Q3=numeric_data.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

Histogram of feature 1 for class A

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data=pd.read_csv("/content/A1.csv.csv")
class_A_data=data[data['Classes']=='A']

plt.figure(figsize=(8,5))
sns.histplot(class_A_data['Feature 1'],bins=10,kde=True,color='pink')
plt.title("Histogram of Feature 1 for Class A")
plt.xlabel('Feature 1')
plt.ylabel('Frequency')
plt.grid(True)
plt.show
```

Box plot for Feature 2 for each class separately

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("/content/A1.csv.csv")

plt.figure(figsize=(8,6))
sns.boxplot(x='Classes',y='Feature 2',data=data)
plt.title("Box plot for Feature 2 for each class")
plt.xlabel('class')
plt.ylabel('Feature 2')
plt.tight_layout()
plt.show()
```

Voilin plot for Feature 3 for each class

Scatter plot between Feature 1 and Feature 3 showing classes separately

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("/content/A1.csv.csv")

print(data.head())

plt.figure(figsize=(8,6))
sns.scatterplot(x='Feature 1',y='Feature 3',hue='Classes',data=data)
plt.title("Scatter plot between Feature 1 and Feature 2")
plt.xlabel('Feature 1')
plt.ylabel('Feature 3')
plt.legend(title='Class')
plt.tight_layout()
plt.show()
```

Contour plot between feature 1 and feature 4 showing classes separately.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("/content/A1.csv.csv")

plt.figure(figsize=(8, 6))
for cls in data['Classes'].unique():
    subset = data[data['Classes'] == cls]
    sns.kdeplot(
        x=subset['Feature 1'],
        y=subset['Feature 4'],
        cmap="cool",
        fill=True,
        thresh=0.05,
        levels=100,
        label=f'Class {cls}'
    )
plt.title('Contour Plot between Feature1 and Feature4 by Class')
plt.xlabel('Feature 1')
plt.ylabel('Feature 4')
plt.legend()
plt.show()
```

Hexagonal bin plot for class A between features 2 and 4.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("/content/A1.csv.csv")

class_a = data[data['Classes'] == 'A']
plt.figure(figsize=(8, 6))
plt.hexbin(class_a['Feature 2'], class_a['Feature 4'], gridsize=30, cmap='viridis')
plt.colorbar(label='Count in bin')
plt.title('Hexbin Plot of Class A: Feature2 vs Feature4')
plt.xlabel('Feature 2')
plt.ylabel('Feature 4')
plt.show()
```

Correlation matrix for the four features

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("/content/A1.csv.csv")

corr = data[['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4']].corr()
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Features')
plt.show()
```

Pair plot for the four features showing classes separately

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("/content/A1.csv.csv")

sns.pairplot(data, vars=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'], hue='Classes', diag_kind='kde')
plt.suptitle('Pair Plot of Features Colored by Class', y=1.02)
plt.show()
```
